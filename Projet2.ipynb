```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Load dataset
df = pd.read_csv("diabetes_prediction_dataset.csv")

# Explore dataset
df.head()

```

```python
# Preprocess data
# Convert categorical variables into numerical variables
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])
df['smoking_history'] = label_encoder.fit_transform(df['smoking_history'])
df['HbA1c_level'] = label_encoder.fit_transform(df['HbA1c_level'])
```

```python
# Separate features and target variable
X = df.drop(['diabetes'], axis=1).values
y = df['diabetes'].values

# Split data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Import different machine learning algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create a dictionary of classifiers
classifier = {'Logistic Regression': LogisticRegression(),
              'Support Vector Machine': SVC(),
              'Decision Tree Classifier': DecisionTreeClassifier(),
              'Random Forest Classifier': RandomForestClassifier()}

# Train and evaluate different classifiers
for name, clf in classifier.items():
    clf.fit(X_train, y_train)
    print(name)
    print("Accuracy on training set: {:.2f}".format(clf.score(X_train, y_train)))
    print("Accuracy on testing set: {:.2f}".format(clf.score(X_test, y_test)))
    print()
```

```python

# Evaluate model using k-fold cross validation
from sklearn.model_selection import cross_val_score

for name, clf in classifier.items():
    scores = cross_val_score(clf, X, y, cv=5) # 5-fold cross validation
    print(name)
    print("Scores: {}".format(scores))
    print("Mean score: {:.2f}".format(scores.mean()))
    print("Standard Deviation: {:.2f}".format(scores.std()))
    print()
```

```python

# Model selected: Random Forest Classifier
# Train final model
final_model = RandomForestClassifier()
final_model.fit(X, y)

```

```python
# Make predictions on new data
patient1 = np.array([1, 50, 0, 1, 2, 170, 7, 270]).reshape(1, -1) ## diabetes = 1
patient2 = np.array([0, 35, 1, 0, 1, 25, 1, 140]).reshape(1, -1) ## no_diabetes = 0

  
```

```python
y1 = final_model.predict(patient1) 
print(y1)
```
